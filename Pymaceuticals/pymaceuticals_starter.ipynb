## Observations and Insights 

```python
# Dependencies and Setup
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
df_mouse_metadata = pd.read_csv(mouse_metadata_path)
df_study_results = pd.read_csv(study_results_path)

df_study_results
```

```python
df_mouse_metadata
```

```python
# Combine the data into a single dataset

df_combined_mouse_data = df_study_results.join(df_mouse_metadata.set_index('Mouse ID'), 
                                               on='Mouse ID')

# Display the data table for preview

df_combined_mouse_data
```

```python
# Checking the number of mice.

df_combined_mouse_data["Mouse ID"].value_counts()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

has_duplicate = df_combined_mouse_data.duplicated(subset=['Timepoint', 'Mouse ID'])
has_duplicate.value_counts()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicates = df_combined_mouse_data[has_duplicate]
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

df_cleaned_mouse_data = df_combined_mouse_data[~df_combined_mouse_data["Mouse ID"].isin(duplicates["Mouse ID"])]

df_cleaned_mouse_data
```

```python
# Checking the number of mice in the clean DataFrame.

df_cleaned_mouse_data["Mouse ID"].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

df_grouped_drugs = df_cleaned_mouse_data.groupby(["Drug Regimen"])

# Create a series for tumor size mean by drug regimen
drug_mean = df_grouped_drugs.mean()["Tumor Volume (mm3)"]
drug_mean.name = "Mean"

# Create a series for tumor size median by drug regimen
drug_median = df_grouped_drugs.median()["Tumor Volume (mm3)"]
drug_median.name = "Median"

# Create a series for tumor size median by drug regimen
drug_var = df_grouped_drugs.var()["Tumor Volume (mm3)"]
drug_var.name = "Variance"

# Create a series for tumor size median by drug regimen
drug_std = df_grouped_drugs.std()["Tumor Volume (mm3)"]
drug_std.name = "Standard Deviation"

# Create a series for tumor size median by drug regimen
drug_sem = df_grouped_drugs.sem()["Tumor Volume (mm3)"]
drug_sem.name = "SEM"



df_summary_stats_by_drug = pd.concat([drug_mean, 
                                      drug_median, 
                                      drug_var,
                                     drug_std,
                                     drug_sem], axis = 1).reset_index()



df_summary_stats_by_drug

```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 

mice_by_drug = df_grouped_drugs['Mouse ID'].nunique()
mice_by_drug.plot(figsize=(6,3), kind='bar')
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.

fig1, ax1 = plt.subplots(figsize = (6,3))
mice_by_drug.plot(kind = 'bar')
plt.ylim(10,30)
plt.title("Number of Mice Per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice")

fig1.tight_layout()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

mice_by_gender = df_cleaned_mouse_data["Sex"].value_counts()
mice_by_gender
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot



fig2, ax2 = plt.subplots(figsize=(5,5))
ax2.pie(mice_by_gender, labels=("Male", "Female"), autopct="%1.1f%%")
plt.title("Mice Gender Breakdown Across Study")

fig2.tight_layout()

```
